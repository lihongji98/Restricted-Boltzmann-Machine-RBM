```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
opt_types = ["cdk", "cdk10", "wcd", "pcd", "wpcd"]
sim_types = ["cos", "F"]

path = f"./distance/{opt_types[0]}_{sim_types[0]}_dhb.npy"
print(path)
```

```python
x_axis = np.arange(1000000)
```

```python
cd1_cos_dw = 0.5 * np.load(f"./distance/{opt_types[0]}_{sim_types[0]}_dw.npy") + 0.5
cd10_cos_dw= 0.5 * np.load(f"./distance/{opt_types[1]}_{sim_types[0]}_dw.npy") + 0.5
wcd_cos_dw = 0.5 * np.load(f"./distance/{opt_types[2]}_{sim_types[0]}_dw.npy") + 0.5
pcd_cos_dw = 0.5 * np.load(f"./distance/{opt_types[3]}_{sim_types[0]}_dw.npy") + 0.5
wpcd_cos_dw= 0.5 * np.load(f"./distance/{opt_types[4]}_{sim_types[0]}_dw.npy") + 0.5

plt.figure(figsize=(15,5))
plt.plot(x_axis, wcd_cos_dw, label='wcd')
plt.plot(x_axis, cd10_cos_dw, label='cd10')
plt.plot(x_axis, pcd_cos_dw, label='pcd')
plt.plot(x_axis, cd1_cos_dw, label='cd1')
plt.plot(x_axis, wpcd_cos_dw, label='wpcd')
plt.legend(loc='upper right')
plt.show()
```
