```python
def draw_KL(dataset, x):
    wcd_record = np.load(f"./KL/"+ dataset +"/wcd_1e3.npy")[:x]
    exact_record = np.load(f"./KL/"+ dataset +"/wcd_exact_1e3.npy")[:x]


    epoch = np.arange(x)

    plt.figure(figsize=(5,3))
    plt.plot(epoch, exact_record, label="exact")
    plt.plot(epoch, wcd_record, label="wcd")
    plt.legend()
    plt.xlim([0, x])
    #plt.ylim([0, 0.4])
    plt.xticks(fontsize=8) 
    plt.yticks(fontsize=8) 
    plt.xlabel('epoch',fontsize=10)
    plt.ylabel('KL',fontsize=10)
    plt.savefig('./graph/'+ dataset +'/'+ dataset + '_exact_wcd_KL.png', dpi=800, bbox_inches='tight')  
    plt.show()
    
    exact_record = np.load(f"./KL/" + dataset+ "/cdk_exact_1e3.npy")[:x]
    cdk_record = np.load(f"./KL/" + dataset + "/cdk_1e3.npy")[:x]
    pt_record = np.load(f"./KL/" + dataset + "/pt_1e3.npy")[:x]

    epoch = np.arange(x)

    plt.figure(figsize=(5,3))
    plt.plot(epoch, exact_record, label="exact")
    plt.plot(epoch, cdk_record, label="cdk")
    plt.plot(epoch, pt_record, label="pt")

    plt.legend()
    plt.xlim([0, x])
    #plt.ylim([0, 0.4])
    plt.xticks(fontsize=8) 
    plt.yticks(fontsize=8) 
    plt.xlabel('epoch',fontsize=10)
    plt.ylabel('KL',fontsize=10)
    plt.savefig('./graph/'+ dataset +'/'+ dataset + '_exact_cdk_pt_KL.png', dpi=800, bbox_inches='tight')  
    plt.show()

    pcd_record = np.load(f"./KL/"+ dataset + "/pcd_1e3.npy")[:x]
    wpcd_record = np.load(f"./KL/" + dataset+ "/wpcd_1e3.npy")[:x]
    exact_record = np.load(f"./KL/" + dataset + "/wpcd_exact_1e3.npy")[:x]


    epoch = np.arange(x)

    plt.figure(figsize=(5,3))
    plt.plot(epoch, exact_record, label="exact")
    plt.plot(epoch, pcd_record, label="pcd")
    plt.plot(epoch, wpcd_record, label="wpcd")
    
    plt.legend()
    plt.xlim([0, x])
    #plt.ylim([0, 0.4])
    plt.xticks(fontsize=8) 
    plt.yticks(fontsize=8) 
    plt.xlabel('epoch',fontsize=10)
    plt.ylabel('KL',fontsize=10)
    plt.savefig('./graph/'+ dataset +'/'+ dataset + '_pcd_wpcd_KL.png', dpi=800, bbox_inches='tight')  
    plt.show()
```

```python
BS3_pt_1 = [78,71,69,69,66,
            71,72,73,73,61,
            65,72,61,61,68,
            67,63,69,58,61,
            64,76,68,68,65]
BS3_pt_2 = [81,74,73,75,76,
           78,70,73,68,66,
           71,72,75,71,69,
           69,68,69,66,60,
           77,75,69,66,61]
BS3_pt_1 =np.array(BS3_pt_1).reshape(5,5) 
BS3_pt_2 =np.array(BS3_pt_2).reshape(5,5) 

BS3_pt = (BS3_pt_1 + BS3_pt_2) / 2 / 1e4
BS3_pt = np.round(BS3_pt, 4) * 14
BS3_pt
```

```python
chain_num = np.array([2,4,6,8,10])
gibbs_sampling = np.array([1,3,5,7,9])
KL = BS3_pt
Gibbs_sampling_num ,Chain_num = np.meshgrid(gibbs_sampling, chain_num)

plt.figure(figsize=(12,7))
ax = plt.axes(projection='3d')
ax.plot_surface(Gibbs_sampling_num, Chain_num, KL, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.view_init(35, 70, 0)  


ax.set_xlabel('Gibbs Sampling Num',fontsize=10)
ax.set_ylabel('Chain Num', fontsize=10) 
ax.set_zlabel('KL Divergence', fontsize=10)

ax.tick_params(axis='x', labelsize=7) 
ax.tick_params(axis='y', labelsize=7)
ax.tick_params(axis='z', labelsize=7)

plt.savefig('PT_KL_BS3.png', dpi=800)  
```
